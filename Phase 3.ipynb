```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test
import seaborn as sns
import os
from lifelines import CoxPHFitter
from itertools import combinations

# Set a professional plotting style for all plots
# Using a light grid style now for better contrast with black text
plt.style.use('seaborn-v0_8-whitegrid') 

# --- Configuration ---
DATA_DIR = '.' # Assuming data files are in the current directory
RESULTS_DIR = 'results'
PLOTS_DIR = os.path.join(RESULTS_DIR, 'multi_omics_analysis') # Reusing this for consistency

# Create directories if they don't exist
os.makedirs(PLOTS_DIR, exist_ok=True)
os.makedirs(os.path.join(PLOTS_DIR, 'survival_analysis'), exist_ok=True) # New dir for survival plots
os.makedirs(os.path.join(RESULTS_DIR, 'summary_tables'), exist_ok=True) # For summary tables
os.makedirs(os.path.join(RESULTS_DIR, 'figures'), exist_ok=True) # For additional figures

# Define filenames for data
# cluster_labels.csv is expected to be in DATA_DIR (current directory) from Phase 2 output
CLUSTER_LABELS_FILE = 'cluster_labels.csv'
SURVIVAL_FILE = 'Survival_SupplementalTable_S1_20171025_xena_sp' # Survival data file

print("Starting Phase 3: Dedicated Survival Analysis based on TME Clusters...")

# --- Initialize DataFrames to None ---
df_annotated = None
surv_df = None

# ----------------------------
# Load Cluster Labels (from Phase 2 output)
# ----------------------------
try:
    df_annotated = pd.read_csv(
        os.path.join(DATA_DIR, CLUSTER_LABELS_FILE),
        index_col=0
    )
    df_annotated["Patient_ID"] = df_annotated.index.str.slice(0, 12)
    print(f"Loaded cluster labels: {df_annotated.shape}")
    print(df_annotated.head())
except FileNotFoundError:
    print(f"Error: Cluster labels file not found at {os.path.join(DATA_DIR, CLUSTER_LABELS_FILE)}. Please ensure Phase 2 has been run and generated this file.")
    df_annotated = None # Set to None on error
except Exception as e:
    print(f"An error occurred loading cluster labels: {e}")
    df_annotated = None # Set to None on error

# ----------------------------
# Load Survival Data (OS)
# ----------------------------
try:
    surv_df = pd.read_csv(os.path.join(DATA_DIR, SURVIVAL_FILE), sep="\t")
    surv_df = surv_df.rename(columns={"_PATIENT": "Patient_ID"})
    # Drop duplicates in survival data based on Patient_ID, keep the first entry
    surv_df = surv_df.drop_duplicates(subset='Patient_ID', keep='first')
    print(f"Loaded survival data: {surv_df.shape}")
    print(surv_df.head())
except FileNotFoundError:
    print(f"Error: Survival file not found at {os.path.join(DATA_DIR, SURVIVAL_FILE)}. Please ensure it's in the data directory.")
    surv_df = None # Set to None on error
except Exception as e:
    print(f"An error occurred loading survival data: {e}")
    surv_df = None # Set to None on error

# ----------------------------
# Merge on Patient ID
# ----------------------------
# Ensure 'cancer_type' and 'cancer type abbreviation' are available if needed later
# For now, only merge what's strictly necessary for the provided survival analysis

# Check if both dataframes are loaded and not empty before merging
if df_annotated is not None and not df_annotated.empty and \
   surv_df is not None and not surv_df.empty:
    merged_df = pd.merge(df_annotated, surv_df, on="Patient_ID", how="inner")
    merged_df = merged_df[["Cluster_Label", "Patient_ID", "OS", "OS.time", "age_at_initial_pathologic_diagnosis", "cancer type abbreviation"]].dropna()
    merged_df["OS"] = merged_df["OS"].astype(int)

    if merged_df.empty:
        print("After merging and dropping NaNs, no common samples with complete survival and cluster data. Skipping all survival analysis.")
    else:
        print(f"\nMerged data for survival analysis shape: {merged_df.shape}")
        print("First 5 rows of merged data:")
        print(merged_df.head())

        # ----------------------------
        # Kaplan–Meier Curve Plot by TME Cluster
        # ----------------------------
        print("\n--- Generating Kaplan-Meier Curve by TME Cluster ---")
        plt.figure(figsize=(10, 7)) # Adjusted figure size
        kmf = KaplanMeierFitter()

        unique_clusters = merged_df["Cluster_Label"].unique()
        if len(unique_clusters) > 1:
            # Define a color palette for the clusters
            palette = sns.color_palette("tab10", n_colors=len(unique_clusters))
            for i, label in enumerate(unique_clusters):
                group = merged_df[merged_df["Cluster_Label"] == label]
                if not group.empty:
                    kmf.fit(durations=group["OS.time"], event_observed=group["OS"], label=label)
                    kmf.plot(ci_show=False, linewidth=2.5, color=palette[i]) # Increased linewidth
            plt.title("Kaplan–Meier Curve: Overall Survival (OS) by TME Cluster", fontsize=16, weight='bold', color='black') # Changed title color to black
            plt.xlabel("Time (days)", fontsize=12, color='black') # Changed label color to black
            plt.ylabel("Survival Probability", fontsize=12, color='black') # Changed label color to black
            plt.xticks(fontsize=10, color='black') # Changed tick color to black
            plt.yticks(fontsize=10, color='black') # Changed tick color to black
            plt.grid(True, linestyle='--', alpha=0.6, color='gray') # Lighter gridlines
            plt.legend(title="TME Cluster", loc="lower left", fontsize=10, title_fontsize=12, frameon=True, facecolor='lightgray', edgecolor='black', labelcolor='black') # Changed labelcolor for legend text to black
            sns.despine(left=True, bottom=True) # Remove top and right spines
            plt.tight_layout()
            plt.savefig(os.path.join(PLOTS_DIR, 'survival_analysis', "km_os_by_cluster.png"), dpi=600, facecolor='white') # Changed facecolor to white
            plt.show() # Display plot
            print(f"Kaplan-Meier curve saved to {os.path.join(PLOTS_DIR, 'survival_analysis', 'km_os_by_cluster.png')}")
        else:
            print("Not enough unique TME clusters to plot Kaplan-Meier curve.")

        # ----------------------------
        # Log-Rank Test Between TME Cluster Groups
        # ----------------------------
        print("\n--- Performing Log-Rank Test (OS) between TME Clusters ---")
        if len(unique_clusters) > 1:
            print("\n📈 Log-Rank Test (OS):")
            for g1, g2 in combinations(unique_clusters, 2):
                d1 = merged_df[merged_df["Cluster_Label"] == g1]
                d2 = merged_df[merged_df["Cluster_Label"] == g2]
                
                # Ensure enough data for log-rank test
                if not d1.empty and not d2.empty:
                    result = logrank_test(d1["OS.time"], d2["OS.time"], event_observed_A=d1["OS"], event_observed_B=d2["OS"])
                    print(f"{g1} vs {g2}: p-value = {result.p_value:.4f}")
                else:
                    print(f"Skipping Log-Rank test for {g1} vs {g2}: Insufficient data.")
        else:
            print("Not enough unique TME clusters to perform Log-Rank test.")

        # ----------------------------
        # Median Survival Table per Cluster
        # ----------------------------
        print("\n--- Calculating Median Overall Survival per Cluster ---")
        medians = (
            merged_df.groupby("Cluster_Label")["OS.time"]
            .median()
            .sort_values(ascending=False)
            .reset_index()
            .rename(columns={"OS.time": "Median_OS_days"})
        )
        print("\n📊 Median Overall Survival per Cluster:")
        print(medians)
        medians.to_csv(os.path.join(RESULTS_DIR, 'summary_tables', "cluster_median_os.csv"), index=False)
        print(f"Median OS per cluster saved to {os.path.join(RESULTS_DIR, 'summary_tables', 'cluster_median_os.csv')}")

        # ----------------------------
        # Multivariate Cox Model (Cluster + Age ONLY)
        # ----------------------------
        print("\n--- Running Multivariate Cox Model (TME Cluster + Age) ---")
        # Select relevant columns
        multi_df = merged_df[[
            "Cluster_Label", "OS", "OS.time", "age_at_initial_pathologic_diagnosis"
        ]].copy()

        # Drop NaNs
        multi_df = multi_df.dropna()

        # One-hot encode cluster labels
        # Ensure there's more than one unique cluster label to encode
        if len(multi_df['Cluster_Label'].unique()) > 1:
            multi_df = pd.get_dummies(multi_df, columns=["Cluster_Label"], drop_first=True)
        else:
            print("Only one unique cluster label found for Cox model (Cluster + Age). Skipping one-hot encoding for Cluster_Label.")
            # If only one cluster, remove the 'Cluster_Label' column as it won't be a useful predictor
            multi_df = multi_df.drop(columns=["Cluster_Label"], errors='ignore')

        # Ensure there are still features to fit the model
        if multi_df.empty or multi_df.shape[0] < 2 or 'OS.time' not in multi_df.columns or 'OS' not in multi_df.columns:
            print("Insufficient data or missing survival columns for Cox model (Cluster + Age). Skipping.")
        else:
            try:
                # Fit Cox model
                cph = CoxPHFitter()
                cph.fit(multi_df, duration_col="OS.time", event_col="OS")

                # Show summary
                print("\n🔹 Cox Model Summary (TME Cluster + Age):")
                cph.print_summary()

                # Save summary
                cph.summary.to_csv(os.path.join(RESULTS_DIR, 'summary_tables', "cox_os_cluster_age_summary.csv"))
                print(f"Cox model summary (Cluster + Age) saved to {os.path.join(RESULTS_DIR, 'summary_tables', 'cox_os_cluster_age_summary.csv')}")

                # Forest Plot
                plt.figure(figsize=(9, 6)) # Adjusted figure size
                # Removed 'color' argument to avoid conflict with 'c'
                cph.plot(hazard_ratios=True, linewidth=2)
                plt.title("Cox Model (TME Cluster + Age) - Overall Survival", fontsize=16, weight='bold', color='black') # Changed title color to black
                plt.xlabel("Hazard Ratio", fontsize=12, color='black') # Changed label color to black
                plt.ylabel("Feature", fontsize=12, color='black') # Changed label color to black
                plt.xticks(fontsize=10, color='black') # Changed tick color to black
                plt.yticks(fontsize=10, color='black') # Changed tick color to black
                plt.axvline(1, color='red', linestyle='--', alpha=0.7, label='No Effect') # Added vertical line at HR=1
                plt.legend(fontsize=10, frameon=True, facecolor='lightgray', edgecolor='black', labelcolor='black') # Changed labelcolor to black
                plt.tight_layout()
                plt.savefig(os.path.join(PLOTS_DIR, 'survival_analysis', "cox_os_cluster_age_only.png"), dpi=600, facecolor='white') # Changed facecolor to white
                plt.show() # Display plot
                print(f"Cox model forest plot (Cluster + Age) saved to {os.path.join(PLOTS_DIR, 'survival_analysis', 'cox_os_cluster_age_only.png')}")
            except Exception as e:
                print(f"Error during Cox Model (Cluster + Age) analysis: {e}")
                import traceback
                traceback.print_exc()

        # ----------------------------
        # Boxplot of OS.time by Cluster
        # ----------------------------
        print("\n--- Generating Boxplot of OS Time by TME Cluster ---")
        plt.figure(figsize=(9, 6)) # Adjusted figure size
        # Added hue and legend=False to address FutureWarning
        sns.boxplot(data=merged_df, x="Cluster_Label", y="OS.time", hue="Cluster_Label", palette="viridis", linewidth=1.5, fliersize=3, legend=False)
        plt.title("Boxplot: Overall Survival Time by TME Cluster", fontsize=16, weight='bold', color='black') # Changed title color to black
        plt.ylabel("Overall Survival Time (days)", fontsize=12, color='black') # Changed label color to black
        plt.xlabel("TME Cluster", fontsize=12, color='black') # Changed label color to black
        plt.xticks(rotation=15, ha='right', fontsize=10, color='black') # Changed tick color to black
        plt.yticks(fontsize=10, color='black') # Changed tick color to black
        plt.grid(axis='y', linestyle='--', alpha=0.6, color='gray')
        sns.despine(left=True, bottom=True)
        plt.tight_layout()
        plt.savefig(os.path.join(PLOTS_DIR, 'survival_analysis', "boxplot_os_time_by_cluster.png"), dpi=600, facecolor='white') # Changed facecolor to white
        plt.show() # Display plot
        print(f"Boxplot of OS time by cluster saved to {os.path.join(PLOTS_DIR, 'survival_analysis', 'boxplot_os_time_by_cluster.png')}")

        # ----------------------------
        # Bar Plot - Median OS per Cluster
        # ----------------------------
        print("\n--- Generating Bar Plot of Median OS per Cluster ---")
        medians_plot = (
            merged_df.groupby("Cluster_Label")["OS.time"]
            .median()
            .reset_index()
            .rename(columns={"OS.time": "Median_OS_days"})
        )
        plt.figure(figsize=(8, 5)) # Adjusted figure size
        # Added hue and legend=False to address FutureWarning
        sns.barplot(data=medians_plot, x="Cluster_Label", y="Median_OS_days", hue="Cluster_Label", palette="magma", edgecolor='black', linewidth=1, legend=False)
        plt.title("Median Overall Survival Time per TME Cluster", fontsize=16, weight='bold', color='black') # Changed title color to black
        plt.ylabel("Median OS Time (days)", fontsize=12, color='black') # Changed label color to black
        plt.xlabel("TME Cluster", fontsize=12, color='black') # Changed label color to black
        plt.xticks(rotation=15, ha='right', fontsize=10, color='black') # Changed tick color to black
        plt.yticks(fontsize=10, color='black') # Changed tick color to black
        plt.grid(axis='y', linestyle='--', alpha=0.6, color='gray')
        sns.despine(left=True, bottom=True)
        plt.tight_layout()
        plt.savefig(os.path.join(PLOTS_DIR, 'survival_analysis', "barplot_median_os_by_cluster.png"), dpi=600, facecolor='white') # Changed facecolor to white
        plt.show() # Display plot
        print(f"Bar plot of median OS by cluster saved to {os.path.join(PLOTS_DIR, 'survival_analysis', 'barplot_median_os_by_cluster.png')}")

        # ----------------------------
        # Stacked Bar of Events (Dead vs Alive)
        # ----------------------------
        print("\n--- Generating Stacked Bar Plot of Events per Cluster ---")
        status_counts = (
            merged_df.groupby(["Cluster_Label", "OS"])
            .size()
            .unstack(fill_value=0)
            .rename(columns={0: "Censored", 1: "Event"})
        )

        if not status_counts.empty:
            ax = status_counts.plot(
                kind="bar", stacked=True, color=["#66c2a5", "#fc8d62"], figsize=(9, 6), # Changed colors, adjusted figsize
                edgecolor='black', linewidth=1 # Added edge color
            )
            plt.title("Stacked Bar: Events vs Censored per TME Cluster", fontsize=16, weight='bold', color='black') # Changed title color to black
            plt.ylabel("Number of Patients", fontsize=12, color='black') # Changed label color to black
            plt.xlabel("TME Cluster", fontsize=12, color='black') # Changed label color to black
            plt.xticks(rotation=45, ha='right', fontsize=10, color='black') # Changed tick color to black
            plt.yticks(fontsize=10, color='black') # Changed tick color to black
            plt.legend(title="Survival Status", loc='upper right', fontsize=10, title_fontsize=12, frameon=True, facecolor='lightgray', edgecolor='black', labelcolor='black') # Changed labelcolor to black
            plt.grid(axis='y', linestyle='--', alpha=0.6, color='gray')
            sns.despine(left=True, bottom=True)
            plt.tight_layout()
            plt.savefig(os.path.join(PLOTS_DIR, 'survival_analysis', "stackedbar_os_events_by_cluster.png"), dpi=600, facecolor='white') # Changed facecolor to white
            plt.show() # Display plot
            print(f"Stacked bar plot of OS events by cluster saved to {os.path.join(PLOTS_DIR, 'survival_analysis', 'stackedbar_os_events_by_cluster.png')}")
        else:
            print("No data for stacked bar plot of events per cluster.")

        # ----------------------------
        # Risk Score Calculation and Visualization
        # ----------------------------
        print("\n--- Calculating and Visualizing Risk Scores ---")
        # Prepare DataFrame for Cox model (for risk score calculation)
        cox_df_risk = merged_df[[
            "Cluster_Label", "OS", "OS.time", "age_at_initial_pathologic_diagnosis"
        ]].copy()

        # Drop rows with NaNs (required for Cox model)
        cox_df_risk = cox_df_risk.dropna()

        # One-hot encode Cluster_Label
        if len(cox_df_risk['Cluster_Label'].unique()) > 1:
            cox_df_risk = pd.get_dummies(cox_df_risk, columns=["Cluster_Label"], drop_first=True)
        else:
            print("Only one unique cluster label found for Cox model (Cluster + Age). Skipping one-hot encoding for Cluster_Label.")
            cox_df_risk = cox_df_risk.drop(columns=["Cluster_Label"], errors='ignore')

        if cox_df_risk.empty or cox_df_risk.shape[0] < 2 or 'OS.time' not in cox_df_risk.columns or 'OS' not in cox_df_risk.columns:
            print("Insufficient data for risk score calculation. Skipping.")
        else:
            try:
                # Fit Cox model
                cph_risk = CoxPHFitter()
                cph_risk.fit(cox_df_risk, duration_col="OS.time", event_col="OS")

                # Calculate risk score
                cox_df_risk["risk_score"] = cph_risk.predict_partial_hazard(cox_df_risk)

                # For grouping, bring back original cluster info
                # Ensure alignment by index
                cox_df_risk["Cluster_Label"] = merged_df.loc[cox_df_risk.index, "Cluster_Label"]

                # Boxplot of risk scores across TME clusters
                plt.figure(figsize=(9, 6)) # Adjusted figure size
                # Added hue and legend=False to address FutureWarning
                sns.boxplot(data=cox_df_risk, x="Cluster_Label", y="risk_score", hue="Cluster_Label", palette="plasma", linewidth=1.5, fliersize=3, legend=False)
                plt.title("Risk Score by TME Cluster", fontsize=16, weight='bold', color='black') # Changed title color to black
                plt.ylabel("Cox Model Risk Score", fontsize=12, color='black') # Changed label color to black
                plt.xlabel("TME Cluster", fontsize=12, color='black') # Changed label color to black
                plt.xticks(rotation=15, ha='right', fontsize=10, color='black') # Changed tick color to black
                plt.yticks(fontsize=10, color='black') # Changed tick color to black
                plt.grid(axis='y', linestyle='--', alpha=0.6, color='gray')
                sns.despine(left=True, bottom=True)
                plt.tight_layout()
                plt.savefig(os.path.join(PLOTS_DIR, 'survival_analysis', "risk_score_boxplot_by_cluster.png"), dpi=600, facecolor='white') # Changed facecolor to white
                plt.show() # Display plot
                print(f"Risk score boxplot by cluster saved to {os.path.join(PLOTS_DIR, 'survival_analysis', 'risk_score_boxplot_by_cluster.png')}")

                # Median-based risk stratification
                median_risk = cox_df_risk["risk_score"].median()
                cox_df_risk["Risk_Group"] = ["High Risk" if r > median_risk else "Low Risk" for r in cox_df_risk["risk_score"]]

                # Kaplan-Meier for High vs Low Risk Groups
                kmf_risk = KaplanMeierFitter()
                plt.figure(figsize=(10, 7)) # Adjusted figure size
                # Define specific colors for High/Low Risk
                risk_palette = {'High Risk': 'red', 'Low Risk': 'green'}
                for group, subdf in cox_df_risk.groupby("Risk_Group"):
                    if not subdf.empty:
                        kmf_risk.fit(subdf["OS.time"], event_observed=subdf["OS"], label=group)
                        kmf_risk.plot(ci_show=False, linewidth=2.5, color=risk_palette.get(group, 'gray')) # Use specific colors
                plt.title("Overall Survival: High vs Low Risk Groups", fontsize=16, weight='bold', color='black') # Changed title color to black
                plt.xlabel("Time (days)", fontsize=12, color='black') # Changed label color to black
                plt.ylabel("Survival Probability", fontsize=12, color='black') # Changed label color to black
                plt.xticks(fontsize=10, color='black') # Changed tick color to black
                plt.yticks(fontsize=10, color='black') # Changed tick color to black
                plt.grid(True, linestyle='--', alpha=0.6, color='gray')
                plt.legend(title="Risk Group", loc="lower left", fontsize=10, title_fontsize=12, frameon=True, facecolor='lightgray', edgecolor='black', labelcolor='black') # Changed labelcolor to black
                sns.despine(left=True, bottom=True)
                plt.tight_layout()
                plt.savefig(os.path.join(PLOTS_DIR, 'survival_analysis', "km_risk_groups.png"), dpi=600, facecolor='white') # Changed facecolor to white
                plt.show() # Display plot
                print(f"Kaplan-Meier plot for high vs low risk groups saved to {os.path.join(PLOTS_DIR, 'survival_analysis', 'km_risk_groups.png')}")

                # Log-rank test for High vs Low Risk Groups
                high_risk_group = cox_df_risk[cox_df_risk["Risk_Group"] == "High Risk"]
                low_risk_group = cox_df_risk[cox_df_risk["Risk_Group"] == "Low Risk"]

                if not high_risk_group.empty and not low_risk_group.empty:
                    result = logrank_test(high_risk_group["OS.time"], low_risk_group["OS.time"], event_observed_A=high_risk_group["OS"], event_observed_B=low_risk_group["OS"])
                    print(f"📈 Log-rank p-value (High vs Low Risk): {result.p_value:.4e}")
                else:
                    print("Cannot perform Log-rank test for High vs Low Risk: One or both groups are empty.")
            except Exception as e:
                print(f"Error during risk score analysis: {e}")
                import traceback
                traceback.print_exc()

        # ----------------------------
        # Multivariate Cox Model: TME Cluster + Age + Cancer Type
        # ----------------------------
        print("\n--- Running Multivariate Cox Model (TME Cluster + Age + Cancer Type) ---")
        # Prepare data
        df_multivariate_cox = merged_df[[
            "Cluster_Label", "OS", "OS.time", "age_at_initial_pathologic_diagnosis", "cancer type abbreviation"
        ]].dropna().copy()

        # Remove rare cancer types for more stable modeling
        cancer_counts = df_multivariate_cox["cancer type abbreviation"].value_counts()
        min_samples_per_cancer_for_cox = 10 # Adjusted threshold for this specific model
        common_cancers = cancer_counts[cancer_counts > min_samples_per_cancer_for_cox].index
        df_multivariate_cox = df_multivariate_cox[df_multivariate_cox["cancer type abbreviation"].isin(common_cancers)]

        if df_multivariate_cox.empty or df_multivariate_cox.shape[0] < 2:
            print("Insufficient data for Multivariate Cox model (Cluster + Age + Cancer Type). Skipping.")
        else:
            # One-hot encoding with drop_first=True
            # Ensure there's more than one unique cluster label and cancer type to encode
            if len(df_multivariate_cox['Cluster_Label'].unique()) > 1:
                df_multivariate_cox = pd.get_dummies(df_multivariate_cox, columns=["Cluster_Label"], drop_first=True)
            else:
                df_multivariate_cox = df_multivariate_cox.drop(columns=["Cluster_Label"], errors='ignore')

            if len(df_multivariate_cox['cancer type abbreviation'].unique()) > 1:
                df_multivariate_cox = pd.get_dummies(df_multivariate_cox, columns=["cancer type abbreviation"], drop_first=True)
            else:
                df_multivariate_cox = df_multivariate_cox.drop(columns=["cancer type abbreviation"], errors='ignore')


            # Drop constant columns (if any) after one-hot encoding
            # This is crucial if some dummy variables end up being constant (e.g., only one cluster/cancer type left)
            df_multivariate_cox = df_multivariate_cox.loc[:, (df_multivariate_cox != df_multivariate_cox.iloc[0]).any()]
            
            # Final check before fitting
            if df_multivariate_cox.empty or df_multivariate_cox.shape[0] < 2 or 'OS.time' not in df_multivariate_cox.columns or 'OS' not in df_multivariate_cox.columns:
                print("Insufficient data or missing survival columns after one-hot encoding and filtering for Multivariate Cox model. Skipping.")
            else:
                try:
                    # Fit Cox model
                    cph_multi = CoxPHFitter() # Corrected typo: CoxPHPHFitter to CoxPHFitter
                    cph_multi.fit(df_multivariate_cox, duration_col="OS.time", event_col="OS")

                    # Show summary
                    print("\n📊 Multivariate Cox Model: TME Cluster + Age + Cancer Type")
                    cph_multi.print_summary()

                    # Plot
                    plt.figure(figsize=(10, 7)) # Adjusted figure size
                    # Removed 'color' argument to avoid conflict with 'c'
                    cph_multi.plot(hazard_ratios=True, linewidth=2)
                    plt.title("Multivariate Cox: Cluster + Age + Cancer Type - Overall Survival", fontsize=16, weight='bold', color='black') # Changed title color to black
                    plt.xlabel("Hazard Ratio", fontsize=12, color='black') # Changed label color to black
                    plt.ylabel("Feature", fontsize=12, color='black') # Changed label color to black
                    plt.xticks(fontsize=10, color='black') # Changed tick color to black
                    plt.yticks(fontsize=10, color='black') # Changed tick color to black
                    plt.axvline(1, color='red', linestyle='--', alpha=0.7, label='No Effect')
                    plt.legend(fontsize=10, frameon=True, facecolor='lightgray', edgecolor='black', labelcolor='black') # Changed labelcolor to black
                    plt.tight_layout()
                    plt.savefig(os.path.join(PLOTS_DIR, 'survival_analysis', "cox_cluster_age_cancer_fixed.png"), dpi=600, facecolor='white') # Changed facecolor to white
                    plt.show() # Display plot
                    print(f"Multivariate Cox model forest plot saved to {os.path.join(PLOTS_DIR, 'survival_analysis', 'cox_cluster_age_cancer_fixed.png')}")

                    # Save results
                    cph_multi.summary.to_csv(os.path.join(RESULTS_DIR, 'summary_tables', "cox_cluster_age_cancer_summary.csv"))
                    print(f"Multivariate Cox model summary saved to {os.path.join(RESULTS_DIR, 'summary_tables', 'cox_cluster_age_cancer_summary.csv')}")
                except Exception as e:
                    print(f"Error during Multivariate Cox (Cluster + Age + Cancer Type) analysis: {e}")
                    import traceback
                    traceback.print_exc()

        # ----------------------------
        # Heatmap: Median OS Across TME Clusters & Cancer Types
        # ----------------------------
        print("\n--- Generating Heatmap of Median OS Across TME Clusters & Cancer Types ---")
        # Group and pivot for heatmap
        heatmap_df = merged_df[["Cluster_Label", "OS.time", "cancer type abbreviation"]].dropna().copy()

        if heatmap_df.empty:
            print("No data for heatmap of median OS across TME clusters and cancer types. Skipping.")
        else:
            # Remove rare cancer types for better visualization in heatmap
            cancer_counts_heatmap = heatmap_df["cancer type abbreviation"].value_counts()
            min_samples_per_cancer_heatmap = 5 # Lower threshold for visualization
            common_cancers_heatmap = cancer_counts_heatmap[cancer_counts_heatmap > min_samples_per_cancer_heatmap].index
            heatmap_df = heatmap_df[heatmap_df["cancer type abbreviation"].isin(common_cancers_heatmap)]

            if heatmap_df.empty:
                print("No sufficient data for heatmap after filtering rare cancer types. Skipping.")
            else:
                pivot_df = (
                    heatmap_df
                    .groupby(["Cluster_Label", "cancer type abbreviation"])["OS.time"]
                    .median()
                    .unstack(fill_value=np.nan) # Use NaN for missing combinations
                )

                # Plot heatmap
                plt.figure(figsize=(18, 7)) # Adjusted figure size
                ax = sns.heatmap(pivot_df, annot=True, fmt=".0f", cmap="coolwarm",
                            linewidths=0.8, linecolor='gray', # Adjusted linewidth, added linecolor
                            cbar_kws={'label': 'Median OS (days)'}, # Removed 'labelcolor' from here
                            mask=pivot_df.isnull(),
                            annot_kws={"size": 10, "color": "black"}) # Changed annotation text color to black
                
                # Manually set colorbar label and tick label colors
                cbar = ax.collections[0].colorbar
                cbar.ax.yaxis.label.set_color('black') # Set colorbar label color to black
                cbar.ax.tick_params(axis='y', colors='black') # Set colorbar tick label colors to black

                plt.title("Median Overall Survival Across TME Clusters & Cancer Types", fontsize=16, weight='bold', color='black') # Changed title color to black
                plt.xlabel("Cancer Type", fontsize=12, color='black') # Changed label color to black
                plt.ylabel("TME Cluster", fontsize=12, color='black') # Changed label color to black
                plt.xticks(rotation=45, ha='right', fontsize=10, color='black') # Changed tick color to black
                plt.yticks(fontsize=10, color='black') # Changed tick color to black
                plt.tight_layout()
                plt.savefig(os.path.join(PLOTS_DIR, 'survival_analysis', "median_os_heatmap.png"), dpi=600, facecolor='white') # Changed facecolor to white
                plt.show() # Display plot
                print(f"Heatmap of median OS across TME clusters and cancer types saved to {os.path.join(PLOTS_DIR, 'survival_analysis', 'median_os_heatmap.png')}")

else:
    print("Skipping survival analysis as either cluster labels or survival data could not be loaded or were empty.")

print("\nPhase 3: Dedicated Survival Analysis complete.")
```
